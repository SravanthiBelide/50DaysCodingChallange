```python
import pandas as pd
```

```python
df = pd.read_csv('help_desk_tickets.csv')
```

```python
df['submission_date'] = pd.to_datetime(df['submission_date'])
df['resolution_date'] = pd.to_datetime(df['resolution_date'])
df['resolution_time_hrs'] = (df['resolution_date'] - df['submission_date']).dt.total_seconds() / 3600
```

```python
df['hour'] = df['submission_date'].dt.hour
df['day_of_week'] = df['submission_date'].dt.day_name()
df['month'] = df['submission_date'].dt.month_name()
volume_by_day = df['day_of_week'].value_counts()

print(df[['submission_date', 'hour', 'day_of_week', 'month']].head())
```

```python
avg_res_total = df['resolution_time_hrs'].mean()
avg_res_by_cat = df.groupby('category')['resolution_time_hrs'].mean()
avg_res_by_priority = df.groupby('priority')['resolution_time_hrs'].mean()
print("Overall Avg Resolution:", avg_res_total)
print("\nAverage by Category:\n", avg_res_by_cat)
print("\nAverage by Priority:\n", avg_res_by_priority)
```

```python
from collections import Counter
import re
all_text = " ".join(df["description"].astype(str))
words = re.findall(r'\b\w+\b', all_text.lower())
word_freq = Counter(words).most_common(20)
print(word_freq)

```

```python
analyst_perf = df.groupby("assigned_analyst")["resolution_time_hrs"].mean().sort_values()
print(analyst_perf)
```
