```python
students = {
    "Asha": {"Math": 78, "Science": 85, "English": 67},
    "Ravi": {"Math": 35, "Science": 40, "English": 50},
    "John": {"Math": 90, "Science": 92, "English": 88}
}
```

```python
results = {}

for name, marks in students.items():
    total = sum(marks.values())
    average = total / len(marks)
    results[name] = {"Total": total, "Average": round(average, 2)}
    
print("Student Stats:", results)
```

```python
top_scorer = max(results, key=lambda x: results[x]["Total"])

print(f"Top Scorer: {top_scorer} with {results[top_scorer]['Total']} marks")
```

```python
failed_students = []

for name, marks in students.items():
    if any(m < 40 for m in marks.values()):
        failed_students.append(name)

print("Failed Students:", failed_students)
```
