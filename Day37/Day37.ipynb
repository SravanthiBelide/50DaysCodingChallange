```python
def get_marks():
    m1 = float(input("Enter marks for Subject 1: "))
    m2 = float(input("Enter marks for Subject 2: "))
    m3 = float(input("Enter marks for Subject 3: "))
    return [m1, m2, m3]
```

```python
def calculate_total(marks):
    return sum(marks)
def calculate_average(total):
    return total / 3
```

```python
def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"
```

```python
student_marks = get_marks()

total_score = calculate_total(student_marks)

average_score = calculate_average(total_score)

final_grade = get_grade(average_score)

print("\n--- Student Report ---")
print(f"Total Marks:   {total_score}")
print(f"Average Marks: {average_score:.2f}")
print(f"Final Grade:   {final_grade}")
```
