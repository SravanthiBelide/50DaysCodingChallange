```python
secret = 7
while True:
    guess = int(input("Guess the number: "))
    if guess < secret:
        print("Too Low")
    elif guess > secret:
        print("Too High")
    else:
        print("You Won!")
        break
```

```python
num = 1234
reversed_num = 0
while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10
print(reversed_num) 
```

```python
num = 56789
count = 0
while num > 0:
    num //= 10 
    count += 1
print(count) 
```

```python
sales = [1200, 800, 1500, 400, 2000]
high_sales_count = 0

for s in sales:
    if s > 1000: 
        print(f"High sale found: {s}")
        high_sales_count += 1 # 

print(f"Total count: {high_sales_count}")
```

```python
total_sum = 0
while True:
    val = int(input("Enter data (0 to stop): ")) # 
    if val == 0:
        break # 
    total_sum += val

print(f"Total Sum: {total_sum}") #
```
