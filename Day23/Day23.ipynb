```python
data = "Ride101-Ramesh-Late Arrival-Bangalore, Ride102-Suresh-rude behavior-Hyderabad, Ride103-Mahesh-Late arrival-Bangalore"
```

```python
clean_data = data.lower()
print(clean_data)
```

```python
total_complaints = clean_data.count("ride")
print(f"Total Complaints: {total_complaints}")
```

```python
late_count = clean_data.count("late arrival")
print(f"Late Arrival Issues: {late_count}")
```

```python
bangalore_count = clean_data.count("bangalore")
print(f"Bangalore Complaints: {bangalore_count}")
```

```python
first_record = clean_data.split(",")[0] 
first_driver = first_record.split("-")[1]
print(f"First Driver Name: {first_driver.capitalize()}")
```

```python
standardized_data = clean_data.replace("rude behavior", "rude_behavior")
standardized_data = standardized_data.replace("late arrival", "late_arrival")
print(standardized_data)
```
