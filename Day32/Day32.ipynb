```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = float(input("Enter a number: "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
n = int(input("Enter N: "))
i = 1
while i <= n:
    print(i)
    i += 1
```

```python
n = int(input("Enter N: "))
total = 0
i = 1
while i <= n:
    total += i
    i += 1
print("Sum:", total)
```

```python
secret = "python123"
attempts = 0
while attempts < 3:
    guess = input("Enter password: ")
    if guess == secret:
        print("Access Granted")
        break
    else:
        attempts += 1
        print("Wrong. Try again.")
```
