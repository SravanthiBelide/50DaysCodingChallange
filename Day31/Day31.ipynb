```python
salary = 50000       
rent = 12000        
food = 8000           
transport = 3000      
entertainment = 4000  
```

```python
# Task 1: Total Expenses
total_expenses = rent + food + transport + entertainment
print(f"Total Expenses: {total_expenses}")
```

```python
# Task 2: Savings Calculation
savings = salary - total_expenses
print(f"Savings: {savings}")
```

```python
# Task 3: Expense Comparison
is_rent_higher = rent > food
is_transport_lower = transport < entertainment
print(f"Is rent greater than food? {is_rent_higher}")
print(f"Is transport less than entertainment? {is_transport_lower}")
```

```python
# Task 4: Financial Condition
if savings > 10000:
    status = "Good Saving"
elif savings >= 5000 and savings <= 10000:
    status = "Average Saving"
else:
    status = "Low Saving"

print(f"Financial Status: {status}")
```

```python
# Task 5: Bonus Update
salary += salary * 0.10  # [cite: 30]
print(f"New Salary after 10% bonus: {salary}")
```
