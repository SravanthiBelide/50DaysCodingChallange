```python
rides = ( 
    (101, "Ravi", "Bangalore", 250, 4.5), 
    (102, "Anjali", "Hyderabad", 180, 4.2), 
    (103, "Kiran", "Bangalore", 300, 4.8), 
    (104, "Meena", "Chennai", 220, 4.0), 
    (105, "Arjun", "Bangalore", 150, 3.9) 
) 
```

```python
total_rides = len(rides)
print(f"Total rides: {total_rides}")
```

```python
bangalore_rides = [ride for ride in rides if ride[2] == "Bangalore"]
print("Bangalore Rides:", bangalore_rides)
```

```python
total_revenue = sum(ride[3] for ride in rides if ride[2] == "Bangalore")
print(f"Total Revenue (Bangalore): {total_revenue}")
```

```python
highest_rated = max(rides, key=lambda ride: ride[4])
print(f"Highest Rated Driver: {highest_rated}")
```

```python
city_data = {}

for ride in rides:
    city = ride[2]
    rating = ride[4]
    
    if city not in city_data:
        city_data[city] = {"total_rating": 0, "count": 0}
    
    city_data[city]["total_rating"] += rating
    city_data[city]["count"] += 1

# Calculate averages
city_averages = {city: round(data["total_rating"] / data["count"], 1) 
                 for city, data in city_data.items()}

print(city_averages)
```
