```python
trips = (
    (101, "Bangalore", 12.5, 250),
    (102, "Hyderabad", 8.0, 180),
    (103, "Bangalore", 15.2, 300),
    (104, "Chennai", 5.5, 120),
    (105, "Bangalore", 7.8, 200)
)
```

```python
total_trips = len(trips)
print(f"Total Trips: {total_trips}")
```

```python
bangalore_count = 0
for trip in trips:
    if trip[1] == "Bangalore":
        bangalore_count += 1
print(f"Trips in Bangalore: {bangalore_count}")
```

```python
total_revenue = sum(trip[3] for trip in trips)
print(f"Total Revenue: {total_revenue}")
```

```python
print("High Distance Trips (> 10km):")
for trip in trips:
    if trip[2] > 10:
        print(trip)
```

```python
highest_fare_trip = max(trips, key=lambda x: x[3])
print(f"Highest Fare Trip: {highest_fare_trip}")
```

```python
bangalore_fares = [trip[3] for trip in trips if trip[1] == "Bangalore"]
if bangalore_fares:
    avg_fare = sum(bangalore_fares) / len(bangalore_fares)
    print(f"Average Fare in Bangalore: {avg_fare}")
```
