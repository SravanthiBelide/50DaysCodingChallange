```python
def calculate_average(values):
    if not values:
        return 0
    return sum(values) / len(values)
```

```python
def find_top_performers(data, column, threshold):
    # Returns a list of students where the score >= threshold
    return [student for student in data if student[column] >= threshold]
```

```python
def find_at_risk_students(data, column, threshold):
    # Returns a list of students where the score < threshold
    return [student for student in data if student[column] < threshold]
```

```python
def find_at_risk_students(data, column, threshold):
    # Returns a list of students where the score < threshold
    return [student for student in data if student[column] < threshold]
```

```python
def batch_summary(data):
    total_students = len(data)
    # Using calculate_average from Step 1
    avg_attendance = calculate_average([s['live'] for s in data])
    avg_sub = calculate_average([s['assignment_sub'] for s in data])
    avg_cons = calculate_average([s['consumption'] for s in data])
    
    return {
        "total_students": total_students,
        "avg_attendance": round(avg_attendance, 1),
        "avg_assignment_submission": round(avg_sub, 1),
        "avg_total_consumption": round(avg_cons, 1)
    }
```

```python
def rank_batches(data):
    # 1. Group data by batch name
    batches = {}
    for student in data:
        name = student['batch']
        if name not in batches:
            batches[name] = []
        batches[name].append(student)
    
    # 2. Analyze each batch
    rankings = {}
    for name, students in batches.items():
        rankings[name] = batch_summary(students)
        
    # 3. Determine best batch (e.g., based on average consumption)
    best_batch = max(rankings, key=lambda b: rankings[b]['avg_total_consumption'])
    return best_batch, rankings
```
