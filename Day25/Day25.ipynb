```python
subjects = ("Math", "Science", "English", "Math", "Science")
total = len(subjects) 
unique = len(set(subjects))
print(f"Total: {total}, Unique: {unique}")
```

```python
feedback = ("  good course  ", "  excellent ", "average ")
clean_feedback = tuple(f.strip() for f in feedback)
print(clean_feedback)
```

```python
names = ("Anu", "Ravi", "John")
courses = ("Python", "SQL", "Power BI")
combined = tuple(f"{n}-{c}" for n, c in zip(names, courses))
print(combined)
```
