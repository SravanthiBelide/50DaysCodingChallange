```python
def calculate_premium(age, policy_type, claims):
    if policy_type == "health":
        base_premium = 8000
    elif policy_type == "vehicle":
        base_premium = 12000
    elif policy_type == "life":
        base_premium = 10000
    else:
        return "Error: Invalid policy type"

    
    if age < 25:
        subtotal = base_premium * 1.20  # Adds 20%
    elif 25 <= age <= 45:
        subtotal = base_premium         # No adjustment
    else:
        subtotal = base_premium * 1.15  # Adds 15%

    if claims == 0:
        final_premium = subtotal * 0.90  # 10% discount
    elif claims > 2:
        final_premium = subtotal * 1.25  # 25% increase
    else:
        final_premium = subtotal         # No adjustment

    return base_premium, final_premium

print("--- Insurance Premium Calculator ---")
user_name = input("Enter customer name: ")
user_age = int(input("Enter age: "))
user_policy = input("Enter policy type (health/vehicle/life): ").lower().strip()
user_claims = int(input("Enter number of claims: "))

# Call the function
result = calculate_premium(user_age, user_policy, user_claims)

# Check if an error occurred or print outputs
if isinstance(result, str):
    print(result)
else:
    base, final = result
    print("\n-------------------------")
    print(f"Customer Name: {user_name}")
    print(f"Policy Type: {user_policy}")
    print(f"Base Premium: ₹{base}")
    print(f"Final Premium: ₹{final:.2f}")
    print("-------------------------")
```
