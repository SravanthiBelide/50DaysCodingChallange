```python
students = ("Anu", "Ravikumar", "Jo")
longest_name = max(students, key=len)
print(longest_name) 
```

```python
subjects = ("Math", "Science", "English")
ui_display = ", ".join(subjects)
print(ui_display) 
```

```python
students = ("Anu", "Ravi")
reversed_names = [name[::-1] for name in students]
print(tuple(reversed_names))
```

```python
students = ("Anu", "Ravi")
vowels = "aeiouAEIOU"

for name in students:
    count = sum(1 for char in name if char in vowels)
    print(f"{name}: {count}") 
```
