```python
order_id = "ORD12345"    
customer_name = "Rahul"  
product_price = 799.99   
quantity = 3             
discount = 0.10          
is_member = True         
```

```python
print(type(order_id))
print(type(customer_name))
print(type(product_price))
print(type(quantity))
print(type(is_member))
```

```python

subtotal = product_price * quantity 
print(f"Check 1 (Subtotal): {subtotal}")

discounted_price = subtotal * (1 - discount)
print(f"Check 2 (After Discount): {discounted_price}")

gst_amount = discounted_price * 0.18
final_bill = discounted_price + gst_amount

print(f"Check 3 (GST Amount): {gst_amount}")
print(f"Check 4 (Final Bill): {final_bill}")
```

```python
print("--- Final Bill ---")
print(f"Order ID: {order_id}") 
print(f"Customer: {customer_name}") 
print(f"Total Price: {subtotal}") 

membership_status = "Yes" if is_member else "No" 
print(f"Discount Applied: {membership_status}")

print(f"Final Amount (including GST): {final_bill:.2f}") 
```
