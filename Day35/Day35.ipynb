```python
nested = [[1, 2], [3, 4], [5]]
flat_list = [item for sublist in nested for item in sublist]
print(flat_list)
```

```python
s1 = "listen"
s2 = "silent"
is_anagram = sorted(s1) == sorted(s2)
print(is_anagram) 
```

```python
nums = [1, 2, 4, 5]
missing = list(set(range(1, 6)) - set(nums))[0]
print(missing) 
```

```python
words = ["apple", "ant", "banana", "bat"]
result = {}
for word in words:
    result.setdefault(word[0], []).append(word)
print(result) 
```

```python
data = {"a": 1, "b": None, "c": 3, "d": None}
filtered_data = {k: v for k, v in data.items() if v is not None}
print(filtered_data) 
```
